Important notes on searching the Unity directory and retrieving data from Pyspark:

2. While downloading data from Spark SQL to Databricks SQL, I encountered issues accessing the data because it isn’t transferred in the standard way. I discovered that the data is incorrectly packaged as a DataFrame—not by the Pandas library, but because Spark has its own functions that operate differently from Pandas. because in this case, to convert my data into a list, I had to collect the scattered data using the `collect()` function, which is not available in the pandas library. So, to summarize, check the Spark documentation first, not the pandas documentation.

```python
crm_table_name = spark.sql("SHOW VOLUMES IN raw_data.dbt_bara_data LIKE 'crm_*'")
crm_table_name = [row['volume_name'] for row in crm_table_name.collect()]

erp_table_name = spark.sql("SHOW VOLUMES IN raw_data.dbt_bara_data LIKE 'erp_*'")
erp_table_name = [row['volume_name'] for row in erp_table_name.collect()]

print(crm_table_name)
print(erp_table_name)
```
